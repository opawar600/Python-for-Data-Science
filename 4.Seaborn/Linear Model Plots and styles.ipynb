```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Linear Model Plot

```python
sns.lmplot(x = 'total_bill', y = 'tip' , data = tips)
#Linear model plot is used to show linear regression in the plot.
```

```python
sns.lmplot(x = 'total_bill', y = 'tip' , data = tips , hue = 'sex')
#Hue is the parameter used to show datapoints on plot with distinct variables present in the given column
```

```python
sns.lmplot(x = 'total_bill', y = 'tip' , data = tips , row = 'smoker' , col = 'time' , hue = 'sex' , markers = ['o','+'])
```

# Styles and color

```python
sns.countplot(x = 'sex' , data = tips)
```

```python
sns.set_style(style = 'darkgrid')
sns.countplot(x = 'sex',data = tips)
```

```python
sns.set_style(style = 'ticks')
sns.countplot(x = 'sex',data = tips)
```

```python
sns.set_context('poster',font_scale=2)
sns.set_style(style = 'darkgrid')
sns.countplot(x = 'sex',data = tips)
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize=(12,7))
sns.countplot(x = 'sex',data = tips)
```

```python
sns.set_context('notebook')
sns.lmplot(x = 'total_bill', y = 'tip' , data = tips , hue = 'sex' , palette='magma')
#For color combination and palette strings, go to seaborn documentation page.
```

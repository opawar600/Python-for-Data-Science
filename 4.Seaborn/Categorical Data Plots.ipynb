```python
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
```

```python
df.tail()
```

```python
df.describe()
```

# Barplot



```python
sns.barplot(df['sex'],df['total_bill'],estimator=np.mean)
#Barplot compares to variables according to the estimator provided
#Here, mean of the total bill is estimator, so it just shows mean of both male and female 
```

# Countplot

```python
sns.countplot(df['sex'])
#Type of barplot which gives the count of all unique values present in the categorical column
```

# Boxplot

```python
sns.boxplot(df['day'],df['tip'])
# Represents the distribution of data in form of quartiles and outliers are shown as well
```

# Violinplot

```python
sns.violinplot(df['day'],df['total_bill'],hue=df['smoker'],split=True)
#Shows the kernel density estimation. hue is used to split the data acoording to values in the specified column
```

# Swarmplot

```python
sns.swarmplot(df['day'],df['total_bill'],hue=df['smoker'])
#Similar to violin plot. A combination of violin and scatterplot
```

# Stripplot

```python
sns.stripplot(df['day'],df['total_bill'],hue=df['sex'])
#Shows how densely the points are located.
```

# Factorplot

```python
sns.factorplot(x='day',y='total_bill',data=df,kind='violin')
sns.factorplot(x='day',y='total_bill',data=df,kind='swarm',color='black')
#Generalised method which can be used to draw any kind of plots
#The 'kind' parameter takes in the kind of plot user wants. It can be bar,swarm,violin,etc
```

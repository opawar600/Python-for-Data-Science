```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
#Dataset is avaiable in seaborn library.
```

```python
tips.head()
```

```python
tips.describe()
```

# Distribution Plot

```python
sns.distplot(tips["total_bill"],kde=False,bins=40)
#Pass the column as first argument. kde is Kernel Density Estimation
```

# Joint Plots

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter")
#Can be used to plot two variables to find the realtion in them
#Here it can be seen that as the bill amount increases, the tip given also increases.
```

# Pairplot

```python
sns.pairplot(tips,hue="sex")
#This plot is used to plot each variable in the dataframe agaist all others
```

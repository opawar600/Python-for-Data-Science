```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,5)
```

```python
y = x**3
```

```python
x
```

```python
y
```

# Functional method for plotting

```python
#Functional method
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Functional method plot")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"b")

plt.subplot(1,2,2)
plt.plot(y,x,"r")

plt.show()
```

# Object Oriented method for plotting

```python
#Object oriented method

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Object Oriented method plotting")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9]) #Empty space for outer plot
axes2 = fig.add_axes([0.19,0.45,0.45,0.45]) #Empty space for inner plot
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9]) #Empty space for outer plot
axes2 = fig.add_axes([0.19,0.45,0.45,0.45]) #Empty space for inner plot


axes1.plot(x,y,"r")
axes1.set_xlabel("X axis")
axes1.set_ylabel("Y axis")
axes1.set_title("Outer Plot")

axes2.plot(y,x)
axes2.set_xlabel("X axis")
axes2.set_ylabel("Y axis")
axes2.set_title("Inner Plot")
```

# Subplots using OOP method

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#Subplots using object oriented method
axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x,"r")
axes[1].set_title("Second Plot")

plt.tight_layout()#Any overlapping of the plots is fixed by this command. Recommended to use this for subplots everytime
```

# Figure Size and DPI

```python
fig = plt.figure(figsize = (4,3)) #Figsize takes in a tuple which specifies the length and height of the figure

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1,figsize = (6,4))

axes[0].plot(x,y)
axes[1].plot(y,x,"r")

plt.tight_layout()
```

```python
#Save an image
fig.savefig("Picture125.png",dpi=200) #dpi is dots per inch. More the number, bigger the picture
```

# Legend

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x*2,label = "X * 2") #Add labels to the respective plots so that they can be shown as a legend on plot
ax.plot(x,x*3,"r",label = "X * 3")

plt.legend(loc="upper left")#Call this method and specify the location where legend should show up
```

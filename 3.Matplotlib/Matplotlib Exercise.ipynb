```python
import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("Title")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1.3,1])
ax2 = fig1.add_axes([0.07,0.55,0.5,0.4])

ax1.plot(x,y,color="red")
ax2.plot(x,y,color="blue")
```

```python
fig3 = plt.figure()

ax11 = fig3.add_axes([0,0,1,1])
ax22 = fig3.add_axes([0.08,0.55,0.6,0.4])

ax11.plot(x,z)
ax11.set_xlabel("X")
ax11.set_ylabel("Y")
ax11.set_title("Title")

ax22.plot(x,y,color = "r")
ax22.set_xlim([20,40])
ax22.set_ylim(0,100)

```

```python
fig,axes = plt.subplots(1,2,figsize=(12,3))

axes[0].plot(x,y,color="blue",lw=3,ls="--")
axes[1].plot(x,z,color="red",lw=3)

plt.tight_layout()
```
